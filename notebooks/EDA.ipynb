```python
nulos = (
    pd.DataFrame({
        'nulos': df.isnull().sum(),
        'porcentaje': (df.isnull().sum() / len(df)) * 100
    })
    .query('nulos > 0')
    .sort_values(by='porcentaje', ascending=False)
    .reset_index()
)

nulos = nulos.rename(columns={'index': 'columna'})
nulos.index = range(1, len(nulos) + 1)

nulos
```
